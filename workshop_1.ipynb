4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure
and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

```python
import time
import numpy as np
import random

size = 1000
A = [[random.random() for _ in range(size)] for _ in range(size)]
B = [[random.random() for _ in range(size)] for _ in range(size)]

start = time.time()

result = [[0 for _ in range(size)] for _ in range(size)]


for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()
t = end - start
print("Python list time:", t)

arr1 = np.random.rand(size, size)
arr2 = np.random.rand(size, size)

start = time.time()
result_np =  arr1 @ arr2
end = time.time()
t = end - start
print("NumPy array time:", t)


```
